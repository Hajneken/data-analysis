```python
import numpy as np 
import matplotlib.pyplot as plt
from random import gauss
```

## Simulated data


```python
alpha = 2
beta = 0.5
sigma = 1 # variance for noise term is used as second parametr of gauss()
N = 200 # number of samples to be generated

x = np.array([gauss(0,1) for i in range(N)]) #generated data 
noise = np.array([gauss(0, sigma) for i in range(N)])

y = np.multiply(alpha, x) + beta + noise # true model => model we want to predict 

plt.figure(figsize=(10, 7))
plt.scatter(x, y)
plt.title 
plt.title("Generated Data")
plt.xlabel('x')
plt.ylabel('y')


x_plot = np.linspace(-3, 3, 100)
y_true = np.multiply(alpha, x_plot) + beta 
plt.plot(x_plot, y_true, 'g') # regression line a.k.a. True model 


z = np.append( x[:, None], np.ones( (N,1)), axis = 1)
A = np.dot(z.transpose(), z) / N #as derived in lecutre 
b = np.dot(y, z) / N
```
